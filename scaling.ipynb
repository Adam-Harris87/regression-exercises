# 1. Apply the scalers we talked about in this lesson to your data and visualize the results for the unscaled and scaled distribution .

```python
# for col in cols:
scaler = sklearn.preprocessing.MinMaxScaler()
# Note that we only call .fit with the training data,
# but we use .transform to apply the scaling to all the data splits.
scaler.fit(train['area'].values.reshape(-1,1))

area_train_scaled = scaler.transform(train['area'].values.reshape(-1,1))
area_validate_scaled = scaler.transform(val['area'].values.reshape(-1,1))
area_test_scaled = scaler.transform(test['area'].values.reshape(-1,1))

plt.figure(figsize=(13, 6))
plt.subplot(221)
plt.hist(train['area'], bins=25, ec='black')
plt.title('area Original')
plt.subplot(222)
plt.hist(area_train_scaled, bins=25, ec='black')
plt.title('area MinMaxScaler Scaled')

scaler = sklearn.preprocessing.StandardScaler()
# Note that we only call .fit with the training data,
# but we use .transform to apply the scaling to all the data splits.
scaler.fit(train['area'].values.reshape(-1,1))

area_train_scaled = scaler.transform(train['area'].values.reshape(-1,1))
area_validate_scaled = scaler.transform(val['area'].values.reshape(-1,1))
area_test_scaled = scaler.transform(test['area'].values.reshape(-1,1))

# plt.figure(figsize=(13, 6))
# plt.subplot(121)
# plt.hist(train['area'], bins=25, ec='black')
# plt.title('area Original')
plt.subplot(223)
plt.hist(area_train_scaled, bins=25, ec='black')
plt.title('area StandardScaler Scaled')
# plt.show()


scaler = sklearn.preprocessing.RobustScaler()
# Note that we only call .fit with the training data,
# but we use .transform to apply the scaling to all the data splits.
scaler.fit(train['area'].values.reshape(-1,1))

area_train_scaled = scaler.transform(train['area'].values.reshape(-1,1))
area_validate_scaled = scaler.transform(val['area'].values.reshape(-1,1))
area_test_scaled = scaler.transform(test['area'].values.reshape(-1,1))

# plt.figure(figsize=(13, 6))
# plt.subplot(121)
# plt.hist(train['area'], bins=25, ec='black')
# plt.title('area Original')
plt.subplot(224)
plt.hist(area_train_scaled, bins=25, ec='black')
plt.title('area RobustScaler Scaled')
plt.show()
```

# 3. Read the documentation for sklearn's QuantileTransformer. Use normal for the output_distribution and apply this scaler to your data. Visualize the result of your data scaling.

```python
scaler = sklearn.preprocessing.QuantileTransformer(output_distribution='normal')
# Note that we only call .fit with the training data,
# but we use .transform to apply the scaling to all the data splits.
scaler.fit(train['area'].values.reshape(-1,1))

area_train_scaled = scaler.transform(train['area'].values.reshape(-1,1))
area_validate_scaled = scaler.transform(val['area'].values.reshape(-1,1))
area_test_scaled = scaler.transform(test['area'].values.reshape(-1,1))

plt.figure(figsize=(13, 6))
plt.subplot(121)
plt.hist(train['area'], bins=25, ec='black')
plt.title('area Original')
plt.subplot(122)
plt.hist(area_train_scaled, bins=25, ec='black')
plt.title('area QuantileTransformer Scaled')
plt.show()
```
